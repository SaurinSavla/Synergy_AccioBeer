```python
mask = ~data['main.disorder'].isin(['Schizophrenia', 'Obsessive compulsive disorder', 'Addictive disorder'])

# Apply the mask to filter the rows
df_filtered = data[mask]
df_filtered.head()
```

```python
# Now, df_filtered contains the rows where 'main.disorder' is not one of the specified disorders
mask = ~df_filtered['specific.disorder'].isin(['Bipolar disorder'])

# Apply the mask to filter the rows
df_filtered = df_filtered[mask]
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming y_corr is your categorical target variable
# Initialize LabelEncoder
label_encoder = LabelEncoder()

data = df_filtered
# Encode the categorical target variable y_corr
data['main.disorder']= label_encoder.fit_transform(data['main.disorder'])

# Get the unique categories and their corresponding encoded integers
categories = label_encoder.classes_
encoded_integers = label_encoder.transform(categories)

# Display the category along with corresponding integers
print("Category \t Encoded Integer")
for category, encoded_integer in zip(categories, encoded_integers):
    print(f"{category} \t\t\t\t\t {encoded_integer}")
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

encoder.fit(data[['sex']])
data['sex']= encoder.transform(data[['sex']])

encoder.fit(data[['specific.disorder']])
data['specific.disorder']= encoder.transform(data[['specific.disorder']])
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

X= data.drop(columns=['main.disorder'], axis=1)
y = data['main.disorder']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=200, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
ada_classifier = AdaBoostClassifier(n_estimators=50, random_state=42)
ada_classifier.fit(X_train, y_train)

# Make predictions
y_pred = ada_classifier.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("AdaBoost Accuracy:", accuracy)
```

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel='linear')  # You can choose different kernels like 'linear', 'rbf', 'poly', etc.

# Train the model
svm_classifier.fit(X_train, y_train)

# Make predictions
y_pred = svm_classifier.predict(X_test)

# Evaluate model accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the model
gb_classifier.fit(X_train, y_train)

# Make predictions
y_pred = gb_classifier.predict(X_test)

# Evaluate model accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
